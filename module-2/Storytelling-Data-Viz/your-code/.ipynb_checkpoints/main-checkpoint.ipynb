#### In the cell below, calculate the percentage of missing values for each column. 

Make a table containing the column name and the percentage of missing values. Print the columns where more than 20% of values are missing. An example of your output looks like:

![Missing Values](missing-values.png)

[This reference](https://stackoverflow.com/questions/51070985/find-out-the-percentage-of-missing-values-in-each-column-in-the-given-dataset) can help you make the missing values table.

```python
def missing_values (data): 
    lst_cols = data.columns
    lst_new_cols = []
    for col in lst_cols: 
        percentage_null = data[col].isnull().sum()/len(data)*100
        if percentage_null > 20: 
            lst_new_cols.append(col)
    return lst_new_cols

print(missing_values(df))
```

## Bonus Challenge 2 - Exploring Data with Correlation Heatmap

Now you have explored data visualizations with certain fields based on common sense. In the dataset there are many other fields that you are not sure whether they are important factors for the sale price. What is the best way to explore those fields without investigating them individually?

Making scatter matrix is not an option here because there are too many fields which makes it extremely time consuming to create scatter matrix. One option you have is to create a heatmap. Heatmaps are much less expensive to create than scatter matrixes. You can use heatmaps to visualize the pairwise correlations between each two variables.

Here is a [reference](https://seaborn.pydata.org/examples/many_pairwise_correlations.html) you can use to learn how to creat the pairwise correlation heatmap. Your heatmap should look like below:

![Corr Heatmap](heatmap.png)

```python

corre = df.corr()
sns.set(style="dark")
mask = np.triu(np.ones_like(corre, dtype=np.bool))
f, ax = plt.subplots(figsize=(11, 9))
cmap = sns.diverging_palette(220, 10, as_cmap=True)

sns.heatmap(corre, mask=mask, cmap=cmap, vmax=1, center=0,
            square=True, linewidths=.5, cbar_kws={"shrink": .5});
```

In your heatmap, you can easily identify the highly correlated (either positively or negatively) variables by looking for the grids with darker colors.

## Challenge 3 - Present Your Stories

Use the following format to present each factor:

```python
#Title: First Factor - SalesPrice & OverallQual
# Here we can see that the OverallQual has a large influence on the price and of course that makes sense because
# If you use more better materials and your finish quality is higher you can ask more money when you sale! 

sns.lmplot(x='OverallQual',y='SalePrice', data=df)
plt.title('Correlation between OverallQual and Salesprice')
plt.show();

```

```python
# Title: Second Factor - SalesPrice & GrLivArea: 

# Here we can say that the SalesPrice of your home increase if your Ground Living Area increase. 
# If you have a bigger livingspace, there's a lot more work on it, more painting more pipelines, electricity,...
# So in that case it make sense to have a bigger price!

df_grliv = df[["Neighborhood", "SalePrice", "GrLivArea"]].groupby(['Neighborhood']).agg({'SalePrice': 'mean', 'GrLivArea': 'sum'})
df_grliv.plot.bar()
```

```python
# Title: Third Factor - SalesPrice & GarageCars | GarageArea: 

# Another thing that effects your price is the GarageCars because how more cars you have in your Garage how bigger
# your house is normally. If you have a larger garage for more cars, you're house is normally bigger and everybody
# knows that if your house is better & in a good condition you pay more. 


sns.lmplot(x='GarageCars',y='SalePrice', data=df)
plt.title('Correlation between Garage and Salesprice')
plt.show();

```

```python
# Title: Fourth Factor - SalesPrice & Neighborhood:

# The sales price is also determined according to the neighborhood. Of course, we must be careful in drawing 
# this conclusion and we may need to look at additional data. 


df_neighborhood = df.groupby(by='Neighborhood').agg({'SalePrice': 'mean'})
df_neighborhood.sort_values(by='SalePrice', ascending=False, inplace=True)
df_neighborhood.plot.bar(); 
```

```python
import pandas as pd
import numpy as np 
import seaborn as sns

```

```python
data = pd.read_csv('/Users/bert/Desktop/dataV2-labs/module-2/Explore Correlations/Data/CDD.csv', sep="\t")
data.head()
```

```python
#Corr but not the default --> because I think these data is not lineair: 

correlation = data.corr(method='spearman')
correlation
```

```python
#Create a heatmap
from matplotlib import pyplot as plt

plt.figure(figsize=(15,8))
heatmap = sns.heatmap(correlation, annot=True, cmap="YlGnBu")
heatmap 
```

```python
"""
First Pair: highly positively correlated: 
ESTIMATED TOTAL COST & NORMALIZED TOTAL COST

"Because if you make a good estimation of your costs the cost to restore it to a normal condition should be the 
 same. But you gonna have always have small differences."

"""

positive = data[['ESTIMATED TOTAL COST', 'NORMALIZED TOTAL COST']]
```

```python
"""
Second Pair: Almost zero: 
MAGNITUDE & FATALITIES

"There is no relationship between those 2 items. It makes sense because you can have a lot of deaths with a small 
 magnitude and no deaths with a huge magnitude. Also the government try to warn for those disasters because they want
 to have as few deaths as possible. Simply there is no relationship between MAGNITUDE & FATALITIES!

"""
zero = data[['MAGNITUDE', 'FATALITIES']]
```

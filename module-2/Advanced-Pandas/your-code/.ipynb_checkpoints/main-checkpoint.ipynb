# Challenge 2 - Advanced Functions

In this part of the lab, we would like to test complex conditions on the entire dataset at once. Let's start by finding the number of rows where the CGPA is higher than 9 and the student has conducted research.

```python
data_2 = pd.read_csv('/Users/bert/Desktop/dataV2-labs/module-2/Advanced-Pandas/Admission_Predict.csv')
data_2.head()
```

Now return all the rows where the CGPA is greater than 9 and the SOP score is less than 3.5. Find the mean chance of admit for these applicants.

```python
filter_data_2 = data_2[(data_2['CGPA']>9) &
                       (data_2['SOP']< 3.5)]

avg_cgpa = round(filter_data_2['CGPA'].mean(), 2)
avg_sop = filter_data_2['SOP'].mean()

print("The mean for CGPA = "+str(avg_cgpa)+" the mean for SOP "+str(avg_sop))
```

We would like to create a deciding factor column for each student. We standardize several columns and then pick the most important factor from a lookup table. If the standardized value is above 0.8, the student will be accepted.

We will start by creating a standardized column for `CGPA`, `GRE Score` and `LOR`. We will name these columns `CGPA_std`, `GRE_std`, and `LOR_std` respecively.

Recall that standardizing a column is done by subtracting the mean of the column from all observations in the column and then dividing each observation in the column by the column's standard deviation.

In the cell below, write a standardization function (a function that takes a column as input and returns a standardized column as output). Make sure to use the `numpy` versions of mean and standard deviation.

```python
def stand_2 (data, col): 
    col_m = np.mean(list(data[col]))
    col_s = np.std(list(data[col]))
    data[col+"_std"] = list((data[col]-col_m)/col_s)
    return data[col+"_std"]
```

Now create the standardized columns `CGPA_std`, `GRE_std`, and `LOR_std` and add them to the `admissions` dataframe.

```python
stand_2(data_2, 'CGPA')
stand_2(data_2, 'GRE Score')
stand_2(data_2, 'LOR ')
data_2.head()
```

We will generate the decision choice at random using the code below. Please run the cell.

```python
# Libraries
from random import choices
```

```python
std_columns = ['CGPA_std', 'GRE Score_std', 'LOR _std']

decision_choice = choices(std_columns, k=data_2.shape[0])
```

Now create the deciding column using the `lookup` function. The lookup column is `decision_choice` found above. Call the column resulting from the lookup function `deciding_column` and add it to the `admissions` dataframe.

```python
data_2['decision_choice'] = data_2.lookup(data_2.index, decision_choice)
```

Create a column called `decision` in the `admissions` dataframe. Assign 1 to this column if the value of `deciding_column` is greater than 0.8 and 0 otherwise.

```python
decision = data_2[data_2['decision_choice'] > 0.8 ]
decision
```

How many applicants will be accepted to the program using the decision column? Compute the result below.

```python
number = round(decision['decision_choice'].sum(), 0)
print("The number of applicants will be "+str(number)+" persons")
```

# Challenge 3 - Method Chaining

To increase our coding efficiency, let's make a number of changes to our dataframe in one line of code.

In the cell below, remove all non character symbols from the column names, replace all spaces in column names to underscores, and change all upper case characters to lower case. Assign these new column values to `admissions.columns`.

```python
lst = data_2.columns 
lst = [col.lower().replace(" ", "_") for col in lst]
data_2.columns = lst


data_3 = data_2
data_3
```

In the cell below, give all student with a university rating of 4 or higher a 10 point boost on their GRE score and split the column into 4 bins using the `cut` function. Assign this new score to the variable `adjusted_gre`.

```python
def gre_increase (row): 
    if row['university_rating'] >= 4: 
        return row['gre_score']+10
    else: 
        return row['gre_score']
    
data_3['gre_score'] = data_3.apply(lambda row: gre_increase(row), axis=1)
```

```python
labels_score = ['Low', 'Average' ,'Good', 'High']
data_3['adjusted_gre'] = pd.cut(data_3['gre_score'],4, labels=labels_score)
data_3.head()
```

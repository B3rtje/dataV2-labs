```python
data.columns = ['code', 'city-part', '2015', '2016', '2017', '2018', '2019',
       '2015.1', '2016.1', '2017.1', '2018.1', '2019.1', '2015.2', '2016.2',
       '2017.2', '2018.2', '2019.2', '2015.3', '2016.3', '2017.3', '2018.3',
       '2019.3']

#Here I changed the column name from Unnamed to 'Code'

data.head(15)
```

```python
data
```

```python
data['12-24'] = data['2015.2']+data['2016.2']+data['2017.2']+data['2018.2']+data['2019.2']

data['25+'] = data['2015.3']+data['2016.3']+data['2017.3']+data['2018.3']+data['2019.3']
```

```python
data.columns
```

```python
data.head(40)
```

```python
data.columns
```

```python

data2 = data[['code', 'city-part', '12-24', '25+']]
```

```python
data_amsterdam_centrum = data2[data2.code == 'A']
```

```python
data_amsterdam_centrum
```

```python
data_amsterdam_centrum['total'] = data_amsterdam_centrum['12-24']+data_amsterdam_centrum['25+']
```

```python
data_amsterdam_centrum
```

```python
#Here we calculate the total number of crimes. 
total_crime = data_amsterdam_centrum['total'].sum()
total_crime
```

```python
data_amsterdam_centrum['percentage'] = data_amsterdam_centrum['total']/total_crime*100
```

```python
data_amsterdam_centrum
```

```python
totale_misdrijven_onder_24 = total_crime - data_amsterdam_centrum['25+'].sum()
totale_misdrijven_onder_24
#Absolute cijfers 

#Percentage 
totale_misdrijven_onder_24_per = totale_misdrijven_onder_24/total_crime*100
#totale_misdrijven_onder_24_per
```

```python
totale_misdrijven_boven_24 = total_crime - data_amsterdam_centrum['12-24'].sum()
totale_misdrijven_boven_24
#Absolute cijfers 

#Percentage 
totale_misdrijven_boven_24_per = totale_misdrijven_boven_24/total_crime*100
totale_misdrijven_boven_24_per
```

```python
#We also wanna know in which street the most crimes happen:

data_streets = data_amsterdam_centrum.sort_values(by=['total'], ascending=False)
data_streets
```

```python
#Here I check if the total crimes equals 100%
data_streets['percentage'].sum()
```

```python
total_labels = ['Extreme Low', 'Low', 'Medium', 'High']

data_streets['labels_data'] = pd.cut(data_streets['total'], 4, labels=total_labels)
data_streets
```

```python
data_streets['each/day'] = (data_streets['total']/5/365).round(2)
data_streets
```

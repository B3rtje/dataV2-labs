```python
import pandas as pd
import numpy as np 
```

```python
data = pd.read_csv('/Users/bert/Desktop/Projecten/amsterdam_p/Misdrijven/soortmisdrijven.csv', sep=';')
```

```python
data.head(15)
```

```python
#First we need to delete the columns with the NaN values. 
null_cols = data.isnull().sum()
null_cols[null_cols > 0]

drop_cols = list(null_cols[null_cols > 10].index)
data = data.drop(drop_cols, axis=1)
data.head(15)
```

```python
#Then we only need the crimes that are commit in the same 2 groups. 

data['12-24'] = data['2015.2']+ data['2016.2']+ data['2017.2']+ data['2018.2'] + data['2019.2']

data['25+'] = data['2015.3']+ data['2016.3']+ data['2017.3']+ data['2018.3'] + data['2019.3']

```

```python
#Then we go to select the columns we need.

data_verwerkt = data[['Soort misdrijf', '12-24', '25+']]
data_verwerkt
```

```python

data_verwerkt['eindtotaal'] = data_verwerkt['12-24'] + data_verwerkt['25+'] 
data_verwerkt
```

```python
#Hoogste criminaliteit bij jongeren
crime = data_verwerkt.sort_values(by=['12-24'], ascending=False)
crime = crime.drop(27)
crime.head(6)

#Here I removed the total. 
```

```python
#Hoogste criminaliteit bij 25 plussers
crime_2 = data_verwerkt.sort_values(by=['25+'], ascending=False)
crime_2 = crime_2.drop(27)
crime_2.head(6)
```

```python
#Making a new names from the original: 

data.rename(columns={'2015.2':'12-24~ 2015',
                     '2016.2':'12-24~ 2016',
                     '2017.2':'12-24~ 2017',
                     '2018.2':'12-24~ 2018',
                     '2019.2':'12-24~ 2019',
                     '2015.3':'+25~ 2015',
                     '2016.3':'+25~ 2016',
                     '2017.3':'+25~ 2017',
                     '2018.3':'+25~ 2018',
                     '2019.3':'+25~ 2019'}, inplace=True)
data.columns

data_evolution = data[['Soort misdrijf','12-24~ 2015', '12-24~ 2016',
       '12-24~ 2017', '12-24~ 2018', '12-24~ 2019', '+25~ 2015', '+25~ 2016',
       '+25~ 2017', '+25~ 2018', '+25~ 2019']]

data_evolution.head()
```

```python
data_evolution['AVG_crimes_year'] = data_evolution[['12-24~ 2015', '12-24~ 2016',
       '12-24~ 2017', '12-24~ 2018', '12-24~ 2019', '+25~ 2015', '+25~ 2016',
       '+25~ 2017', '+25~ 2018', '+25~ 2019']].mean(axis=1)

data_evolution.sort_values('AVG_crimes_year', ascending=False)
```

```python
#Is there a evolution in: Winkeldiefstal, mishandeling, drugshandel, onder invloed
winkeldiefstal = data_evolution[data_evolution['Soort misdrijf'] == 'Winkeldiefstal']
winkeldiefstal
```

```python
lst_crime = ['Winkeldiefstal', 'Mishandeling', 'Drugshandel', 'Onder invloed (weg)']
high_crimes = data_evolution[data_evolution['Soort misdrijf'].isin(lst_crime)]
high_crimes = high_crimes.sort_values(by=['AVG_crimes_year'], ascending=False)
high_crimes
```

```python
high_crimes['high_crimes_jongeren_per'] = (high_crimes['12-24~ 2019'] / high_crimes['12-24~ 2015']-1)*100
```

```python
high_crimes['high_crimes_ouderen_per'] = (high_crimes['+25~ 2019'] / high_crimes['+25~ 2015']-1)*100 
```

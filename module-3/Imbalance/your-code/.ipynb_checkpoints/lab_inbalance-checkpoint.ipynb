# Inbalanced Classes
## In this lab, we are going to explore a case of imbalanced classes. 


Like we disussed in class, when we have noisy data, if we are not careful, we can end up fitting our model to the noise in the data and not the 'signal'-- the factors that actually determine the outcome. This is called overfitting, and results in good results in training, and in bad results when the model is applied to real data. Similarly, we could have a model that is too simplistic to accurately model the signal. This produces a model that doesnt work well (ever).

### Run a logisitc regression classifier and evaluate its accuracy.

```python
#Normally I would scale some numbers but I don't know if we're allowed to do that. 
```

```python
from sklearn.linear_model import LogisticRegression
from sklearn.model_selection import train_test_split

col_names = [col for col in data.columns.values if col not in ['isFraud']]
X = data[col_names]
y = data['isFraud']

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, shuffle=True)
```

```python
lr_model = LogisticRegression().fit(X_train, y_train)
```

```python
y_pred_lr = lr_model.predict(X_test)
```

```python
from sklearn.metrics import confusion_matrix
confusion_lr = confusion_matrix(y_test, y_pred_lr)
print(confusion_lr)
print((1269575+624)/(1269575+624+1301+1024))
```

### Now pick a model of your choice and evaluate its accuracy.

```python
count_class_0, count_class_1 = data.isFraud.value_counts()
print(count_class_0)
print(count_class_1)
```

```python
data_fraud = data[data['isFraud'] == 1]
data_no_fraud = data[data['isFraud'] == 0]
```

```python
df_class_0_under = data_no_fraud.sample(count_class_1)
data_2 = pd.concat([df_class_0_under, data_fraud], axis=0)

print('Random under-sampling:')
print(data_2.isFraud.value_counts())

X2 = data_2[col_names]
y = data_2['isFraud']

X2_train, X2_test, y2_train, y2_test = train_test_split(X2, y, test_size=0.2, shuffle=True)
```

```python
lr_model_2 = LogisticRegression().fit(X2_train, y2_train)
```

```python
y_pred_lr_2 = lr_model_2.predict(X2_test)
```

```python
confusion_lr_2 = confusion_matrix(y2_test, y_pred_lr_2)
print(confusion_lr_2)
print((1502+1482)/(1502+1482+173+129))
```
